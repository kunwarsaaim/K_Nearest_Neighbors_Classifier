```python
from sklearn.datasets import load_digits
```

```python
data_set = load_digits()
```

```python
data_set.keys()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
x_train,x_test,y_train,y_test = train_test_split(data_set['data'],data_set['target'])  #Split data into training and testing
```

```python
max_acc = 0
k=0   # Trying different values of number of Neighbors and saving it in k
for i in range(1,50):
    neigh = KNeighborsClassifier(i)
    model = neigh.fit(x_train,y_train)
    y_prediction = model.predict(x_test)
    acc = accuracy_score(y_test,y_prediction)*100
    if(max_acc<acc):
        max_acc = acc
        k=i
print(k)
    
    
```

```python
l=1
r=999
flag = False
k=0 
#Applying Binary Search  to get optimal value of k
while(l<=r and not flag):
    mid = int((l+r)/2)
    neigh = KNeighborsClassifier(l)
    model = neigh.fit(x_train,y_train)
    y_prediction = model.predict(x_test)
    acc_l = accuracy_score(y_test,y_prediction)*100
    neigh = KNeighborsClassifier(r)
    model = neigh.fit(x_train,y_train)
    y_prediction = model.predict(x_test)
    acc_r = accuracy_score(y_test,y_prediction)*100
    neigh = KNeighborsClassifier(mid)
    model = neigh.fit(x_train,y_train)
    y_prediction = model.predict(x_test)
    acc_mid = accuracy_score(y_test,y_prediction)*100
    print(acc_mid,mid)
    print(acc_l,l)
    print(acc_r,r)
    if(acc_mid>acc_l and acc_mid>acc_r):
        k=mid
        flag = True
    else:
        if(acc_r>acc_mid):
            l = mid
        else:
            r = mid
    
    
    
    
```
